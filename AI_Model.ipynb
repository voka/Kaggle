# 인공지능
인공지능(Artificial Intelligence)는 인간의 지능이 갖고 있는 기능을 갖춘 컴퓨터 시스템을 뜻하며, 인간의 지능을 기계 등에 인공적으로 구현한 것을 말합니다.

머신러닝(Machine Learning) 혹은 기계학습은 인공지능의 한 분야로, 컴퓨터가 학습할 수 있도록 하는 알고리즘과 기술을 개발하는 분야를 뜻합니다.

딥러닝(Deep Learning)은 여러 비선형 변환기법의 조합을 통해 높은 수준의 추상화(다량의 복잡한 자료들에서 핵심적인 내용만 추려내는 작업)을 시도하는 기계학습 알고리즘의 집합으로 뜻합니다.

아래 그림에서 보면 커다란 인공지능의 분야 안에 머신러닝 그 안에 딥러닝이 포함된 것을 알 수 있습니다.

# ANN(Artificial Neural Network)

인공신경망으로 불리는 ANN은 사람의 신경망 원리와 구조를 모방해 만든 기계학습 알고리즘입니다.

- 신경망은 다수의 데이터를 받는 입력층(Input), 출력을 담당하는 출력층(Output), 입력층과 출력층 사이에 존재하는 레이어들(은닉층)이 존재합니다. 
- 여기서 히든 레이어들의 갯수와 노드의 개수를 구성하는 것을 모델을 구성한다고 하는데, 이 모델을 잘 구성하여 원하는 Output값을 잘 예측하는 것이 우리가 해야할 일인 것입니다. 은닉층에서는 활성화함수를 사용하여 최적의 Weight와 Bias를 찾아내는 역할을 합니다.

하지만 이런 ANN은 문제가 있습니다.

- 학습과정에서 파라미터의 최적값을 찾기 어렵습니다.


# DNN(Deep Neural Network)

DNN은 ANN의 은닉층의 개수(2개이상)를 늘려 학습의 결과를 향상시키는 방법입니다.

FCNN(Fully Convolutional neural network)도 딥러닝 모델의 한 종류입니다.

하지만, 이런 DNN도 아래와 같은 문제가 있었습니다.

- Overfitting -> Training Data에만 적합한 해답을 가지고 있어서 일반화(Generalization)하기가 어렵습니다. 
- 모델의 복잡도(Power)가 높으면 Train_data를 따라간는 경향이 있습니다.
- 이렇게 일반화가 어렵다 보니 융통성 없습니다.-> Train dataset에서 조금만 달라져도 완전 다른 결과를 도출해버립니다.
- 은닉층이 많아지고, 데이터가 많을 수록 정확성은 높아질 수 있지만 그만큼 학습하는데 있어 시간이 많이 소요됩니다.

이를 해결하기 위해 나온 모델들이 CNN과 RNN등입니다.

# CNN(Convolution Nerual Network)

FCNN의 은닉층에서 몇개의 노드와 weight를 Drop(제거)함으로써 일반화를 할 수 있도록 만들어진 Model입니다.
또, FCNN은 3차원 데이터를 1차원으로 평면화 시켜야 하기 때문에 공간정보가 손실되는 일이 많았습니다. 이런 정보를 유지한 상태로 학습이 가능한 모델이 CNN입니다.

아래의 항목은 CNN이 가지는 FCNN와 비교되는 차별성 입니다. 
- 각 레이어의 입출력 데이터의 형상 유지
- 이미지의 공각정보를 유지하면서 인접 이미지와 특징을 효과적으로 인식
- 복수의 filter로 이미지의 특징 추출 및 학습
- Pooling Layer 사용 -> 추출한 이미지의 특징을 모으고 강화.
- Filter를 공유 Parameter로 사용하므로 일반적은 DNN과 비교했을 때 학습 parameter가 매우 적다.

이때, 합성곱에 쓰이는 필터는 보통 32, 64개로 설정을 하는데 이걸 이렇게 설정하는 이유는 모르겠지만 많을수록 Overfitting이 될 것이고, 적을 수록 학습이 잘 안될 것입니다. 그리고 합성곱을 거친 결과를 ReLU(Rectified linear unit)이라는 활성화 함수에 통과시킵니다. 
이때, 모델이 점점 딥한 학습을 하게 된다면(레이어가 깊어진다면) Gradient(기울기)가 0에 가까워져 (다양한 결과값이) 전달이 잘 안돼서 학습이 한되는 vanishing gradient문제가 발생할 가능성이 높아집니다. 이를 해결하기 위해 ReLu를 쓰게 되었다고 합니다.
